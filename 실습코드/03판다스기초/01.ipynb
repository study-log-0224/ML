```python
import pandas as pd
menu = pd.Series(['비빔밥', '김치찌개', '된장찌개'])
price = pd.Series([10000, 9000, 8000])
#menu
price
```

```python
df_from_series = pd.DataFrame({'메뉴': menu, '가격': price})
df_from_series
```

```python
df = pd.DataFrame({
    '메뉴':  ['비빔밥', '김치찌개', '된장찌개'],
    '가격':  [10000, 9000, 8000],
    '원산지': ['국내산', '국내산', '국내산']
})
df
```

```python
menu_series = df['메뉴']     # Series
#menu_series
menu_df = df[['메뉴']]       # DataFrame (대괄호 2개)
menu_df
```

```python
cols = ['메뉴', '가격']
#df[cols]
df[['메뉴', '가격']]
```

```python
print('df          :', type(df))
print("df['가격']   :", type(df['가격']))     # Series
print("df[['가격']] :", type(df[['가격']]))   # DataFrame
```

```python
#print(df.loc[2])          # 인덱스 0 행 전체
df.loc[df['가격'] >= 9000, ['메뉴', '원산지']]  # 가격 9000 이상
```

```python
print(df.iloc[0])         # 0번째 행
df.iloc[0:2, 0:2]        # 앞 2행, 앞 2열
```

```python
df['할인가'] = (df['가격'] * 0.9).astype(int)   # 10% 할인 열
df
```

```python
df.loc[df['메뉴'] == '김치찌개', '가격'] = 9500   # 김치찌개 가격 변경
df
```

```python
df = df.drop(columns=['할인가'])
df
```

```python
df.sort_values(by='가격', ascending=False)   # 가격 내림차순
```
