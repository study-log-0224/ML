<details><summary>정답 보기</summary>

```python
raw = ['10', '3.5', 'abc', '7', '', '2.25', None, '100', 'nan']

nums = []
for r in raw:
    if r is None:              # None 먼저 걸러야 아래 메서드에서 에러 안 남
        continue
    if r.isdigit():            # '10' → 정수
        nums.append(int(r))
    elif r.replace('.', '', 1).isdigit():   # '3.5' → 점 1개만 지우고 숫자면 실수
        nums.append(float(r))
    # 그 외('abc', '', 'nan')는 무시

int_cnt   = len([n for n in nums if type(n) is int])
float_cnt = len([n for n in nums if type(n) is float])
print(nums)
print(f'int {int_cnt}개 / float {float_cnt}개')
```

**함정**
- `None`을 먼저 걸러내지 않으면 `None.isdigit()`에서 `AttributeError`.
- `'3.5'.isdigit()`는 **False**입니다 (점 때문). 그래서 점을 1개만 제거해서 검사.
- `type(True) is int` → False지만 `isinstance(True, int)` → **True**입니다. bool은 int의 자식이라서요.

```
이름        점수     비율   등급
------------------------------
홍길동        92   30.7%      A
김철수         8    2.7%      F
이영희       123   41.0%      A
```

- 이름: 왼쪽 정렬 8칸 → `{name:<8}`
- 점수: 오른쪽 정렬 5칸 → `{score:>5}`
- 비율: 소수점 1자리 + `%` → `{ratio:>6.1f}%`  (비율 = 점수 / 전체합계 × 100)
- 등급: 오른쪽 정렬 7칸 (90↑ A, 80↑ B, 70↑ C, 그 외 F)

````python
rows = [('홍길동', 92), ('김철수', 8), ('이영희', 123), ('박민수', 77)]

# TODO: 위 형식대로 출력하세요
````

<details><summary>정답 보기</summary>

```python
rows = [('홍길동', 92), ('김철수', 8), ('이영희', 123), ('박민수', 77)]

total = 0
for name, score in rows:      # 튜플 언패킹
    total += score

def get_grade(s):
    if s >= 90: return 'A'
    elif s >= 80: return 'B'
    elif s >= 70: return 'C'
    return 'F'

print(f"{'이름':<8}{'점수':>5}{'비율':>7}{'등급':>7}")
print('-' * 30)
for name, score in rows:
    ratio = score / total * 100
    print(f'{name:<8}{score:>5}{ratio:>6.1f}%{get_grade(score):>7}')
print('-' * 30)
print(f"{'합계':<8}{total:>5}{100.0:>6.1f}%")
```

**포맷 스펙 정리**
| 표기 | 뜻 |
|---|---|
| `{x:<8}` | 왼쪽 정렬, 8칸 |
| `{x:>5}` | 오른쪽 정렬, 5칸 |
| `{x:^10}` | 가운데 정렬 |
| `{x:.2f}` | 소수점 2자리 |
| `{x:,}` | 천 단위 콤마 (1,234,567) |
| `{x:>8,.1f}` | 조합도 가능 |

주의: 한글은 폭이 2칸이라 터미널에서 살짝 어긋나 보일 수 있습니다. 정상입니다.

<details><summary>정답 보기</summary>

```python
text = "the quick brown fox jumps over the lazy dog. The dog barks, and the fox runs!"

freq = {}
for w in text.lower().split():
    w = w.strip('.,!?')          # 양끝 구두점 제거
    if w == '':
        continue
    freq[w] = freq.get(w, 0) + 1   # 없으면 0에서 시작 → +1

# 값(빈도) 기준 내림차순 정렬
ranked = sorted(freq.items(), key=lambda x: x[1], reverse=True)

for word, cnt in ranked[:3]:      # 상위 3개 (튜플 언패킹)
    print(f'{word:<8}{cnt}회')
```

출력:
```
the     4회
fox     2회
dog     2회
```

**핵심 3가지**
- `d.get(key, 0) + 1` — 카운팅의 정석. `d[key] += 1`은 키가 없으면 **KeyError**.
- `d.items()` → `[('the', 4), ('fox', 2), ...]` 형태의 (키, 값) 튜플 목록.
- `key=lambda x: x[1]` — "각 튜플의 1번(값)을 기준으로 정렬해라"라는 뜻. `lambda`는 이름 없는 1줄짜리 함수입니다.

```python
        result = 'F (출석미달)'
    elif score >= 90 and attendance >= 90:
        result = 'A+'
```

```python
def collatz_steps(n):
    steps = 0
    while n != 1:                 # 1이 되면 멈춤
        if n % 2 == 0:
            n = n // 2            # 주의: / 가 아니라 // (정수 나눗셈)
        else:
            n = 3 * n + 1
        steps += 1
    return steps

best_n, best_steps = 1, 0
for i in range(1, 31):            # 1 ~ 30
    s = collatz_steps(i)
    if s > best_steps:            # 더 크면 갱신
        best_n, best_steps = i, s   # 튜플 동시 대입

print(f'최대: n={best_n}, {best_steps}단계')   # n=27, 111단계
```

<details><summary>정답 보기</summary>

```python
def describe(*nums, ndigits=2, drop_outlier=False):
    values = list(nums)                 # 튜플 → 리스트 (수정 가능하게)

    if drop_outlier and len(values) > 2:
        values.remove(min(values))      # 최솟값 1개 제거
        values.remove(max(values))      # 최댓값 1개 제거

    if len(values) == 0:
        return (0, None, None, None)    # 다중 반환 = 튜플

    total = 0
    for v in values:
        total += v
    avg = round(total / len(values), ndigits)

    return (len(values), avg, min(values), max(values))


# 사용 & 언패킹
cnt, avg, lo, hi = describe(1, 2, 3, 100)
print(cnt, avg, lo, hi)                      # 4 26.5 1 100
print(describe(1, 2, 3, 100, drop_outlier=True))   # (2, 2.5, 2, 3)
```

**핵심**
- `*nums` → 인자를 몇 개 넘기든 **튜플**로 묶여 들어옵니다.
- `ndigits=2` 같은 **기본값 인자**는 반드시 `*nums` 뒤에 오고, 호출 시 `ndigits=0`처럼 **이름으로** 지정합니다.
- `return (a, b, c)`는 튜플 하나를 돌려주는 것. 받을 때 `a, b, c = ...`로 **언패킹**합니다.
- `round(2.0, 0)` → `2.0` (float 유지). 정수로 원하면 `int()`.

```python
# 1) 평탄화 — for가 왼쪽부터 바깥 루프
flat = [v for row in matrix for v in row]

# 2) 전치 — 열 인덱스 j를 바깥, 행 인덱스 i를 안쪽으로
transposed = [[row[j] for row in matrix] for j in range(len(matrix[0]))]

# 3) 조건부 표현식 — for 앞에 오는 if/else (필터링 if와 위치가 다름!)
fizz = ['짝' if v % 2 == 0 else '홀' for row in matrix for v in row]

# 4) 대각선 — enumerate로 인덱스와 값을 동시에
diag = [row[i] for i, row in enumerate(matrix)]

# 5) 데카르트 곱
pairs = [(l, n) for l in letters for n in numbers]
```

**가장 헷갈리는 부분: `if`의 두 위치**

| 형태 | 위치 | 역할 |
|---|---|---|
| `[x for x in L if x > 0]` | **for 뒤** | **필터** (조건 맞는 것만 남김) |
| `['+' if x > 0 else '-' for x in L]` | **for 앞** | **변환** (모든 값을 둘 중 하나로) |

for 앞에 쓸 때는 `else`가 **필수**입니다 (표현식이라 값을 꼭 내놔야 함).

중첩 for의 순서: `for row in matrix for v in row` 는 아래와 같습니다.
```python
for row in matrix:
    for v in row:
        ...
```
왼쪽이 바깥 루프 — 일반 for문을 그대로 쓴 순서라고 기억하세요.

```python
def is_prime(n):
    if n < 2:
        return False
    i = 2
    while i * i <= n:        # 제곱근까지만 확인하면 충분
        if n % i == 0:       # 나누어떨어지면 소수 아님

        i += 1
    return True

twin = [(a, b) for a, b in zip(primes, primes[1:]) if b - a == 2]

print(twin)
```

```python
def get_grade(s):
    if s >= 90: return 'A'
    elif s >= 80: return 'B'
    elif s >= 70: return 'C'
    return 'F'

# --- 유효한 행만 걸러내기 (None 점수 + 키 오타 행 제외)
clean = [r for r in records if r.get('이름') is not None and r.get('점수') is not None]

# --- 1) 과목별: 합계와 개수를 따로 모은 뒤 나눔
sub_sum, sub_cnt = {}, {}
for r in clean:
    s = r['과목']
    sub_sum[s] = sub_sum.get(s, 0) + r['점수']
    sub_cnt[s] = sub_cnt.get(s, 0) + 1
subject_avg = {s: round(sub_sum[s] / sub_cnt[s], 1) for s in sub_sum}   # 딕셔너리 컴프리헨션

# --- 2) 학생별 (같은 패턴)
stu_sum, stu_cnt = {}, {}
for r in clean:
    n = r['이름']
    stu_sum[n] = stu_sum.get(n, 0) + r['점수']
    stu_cnt[n] = stu_cnt.get(n, 0) + 1
student_avg = {n: round(stu_sum[n] / stu_cnt[n], 1) for n in stu_sum}

# --- 3) 1등
ranked = sorted(student_avg.items(), key=lambda x: x[1], reverse=True)
top = ranked[0]

# --- 4) 등급 분포
grade_dist = {}
for n, avg in student_avg.items():
    g = get_grade(avg)
    grade_dist[g] = grade_dist.get(g, 0) + 1

# --- 5) 랭킹 표
print(f"{'순위':<5}{'이름':<8}{'평균':>7}{'등급':>6}")
print('-' * 28)
for i, (name, avg) in enumerate(ranked, start=1):    # 1부터 번호 매기기
    print(f'{i:<5}{name:<8}{avg:>7.1f}{get_grade(avg):>6}')

print()
print('과목별 평균:', subject_avg)
print('등급 분포  :', grade_dist)
```

출력:
```
순위   이름        평균    등급
----------------------------
1    이영희      97.5     A
2    홍길동      90.0     A
3    김철수      71.3     C
4    박민수      62.3     F
```

**이 문제의 함정 3가지**
1. `'이름'` 키가 없는 행 → `r['이름']`은 KeyError. `r.get('이름')`으로 안전 조회 후 필터링.
2. `점수`가 `None`인 행 → 합계에 더하면 `TypeError`. **개수에서도 빼야** 홍길동 평균이 90.0이 됩니다 (270/3=60이 아니라 180/2=90).
3. `홍길동` 평균은 `(92+88)/2 = 90.0`. 결측을 0으로 채우면 60.0이 되어 완전히 다른 결론이 나옵니다 — 실전 ML에서 **결측치 처리**가 왜 중요한지 보여주는 지점입니다.

## 정리 — 반드시 챙길 관용구

| 패턴 | 코드 |
|---|---|
| 카운팅 | `d[k] = d.get(k, 0) + 1` |
| 값 기준 정렬 | `sorted(d.items(), key=lambda x: x[1], reverse=True)` |
| 리스트 뒤집기 | `lst[::-1]` |
| 리스트 복사 | `lst[:]` 또는 `lst.copy()` (`=`는 복사 아님!) |
| 인덱스+값 동시 순회 | `for i, v in enumerate(lst):` |
| 두 리스트 짝짓기 | `for a, b in zip(l1, l2):` |
| 이웃 쌍 만들기 | `zip(lst, lst[1:])` |
| 필터 컴프리헨션 | `[x for x in L if cond]` (if가 **뒤**) |
| 변환 컴프리헨션 | `[A if cond else B for x in L]` (if가 **앞**, else 필수) |
| 평탄화 | `[v for row in M for v in row]` |
| 다중 반환 | `return (a, b)` → `a, b = f()` |
| 정수 나눗셈 | `//` (`/`는 항상 float) |

**주피터 팁**
- `Shift + Enter` 실행 / `Esc` + `A`·`B` 위·아래 셀 추가 / `Esc` + `DD` 셀 삭제
- 무한 루프에 빠지면 상단 **■ (Interrupt)** 버튼
- 셀 실행 순서에 따라 변수가 남아 있습니다. 헷갈리면 **Kernel → Restart & Run All**
- 함수 사용법이 궁금하면 `sorted?` 또는 `Shift + Tab`
