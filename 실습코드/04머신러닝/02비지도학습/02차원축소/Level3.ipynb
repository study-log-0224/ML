```python
# ############################################################
# Level 3 — 와인 품질 데이터 11차원 -> 2차원 (공개 데이터)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기 — 표 / 전처리 / PCA
# ------------------------------------------------------------
import pandas as pd                               # 표(CSV) 다루기
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# ------------------------------------------------------------
# [목적] 데이터 불러오기(세미콜론 구분) + 정답 제외 + 스케일링
# ------------------------------------------------------------
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
df = pd.read_csv(url, sep=';')                    # 이 파일은 ';'(세미콜론)으로 열 구분
X = df.drop(columns='quality')                    # 품질(정답)만 빼고 특성 11개
X_s = StandardScaler().fit_transform(X)           # 스케일링 (PCA 필수 전처리)
```

```python
# ------------------------------------------------------------
# [목적] 11차원 -> 2차원으로 줄이고 지킨 정보 확인
# ------------------------------------------------------------
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X_s)                     # 11열 -> 2열
print('2축이 지킨 정보:', round(pca.explained_variance_ratio_.sum(), 2))
```

```python
# ============================================================
# [결과 해석]
#  · 11 -> 2차원으로 줄이니 지킨 정보 ~ 0.46 (절반 이하)
#  · iris(0.97)보다 낮음 -> 와인 품질 특성들은 서로 덜 겹쳐서 2축만으론 부족
#  · 이럴 땐 축을 더 늘리거나(예: 정보 95%), 원래 특성을 그대로 쓰는 게 나을 수 있음
# ============================================================
```
