```python
# ############################################################
# Level 2 — 30개 특성 중 몇 축이면 정보 95%? (스케일링 필수)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기 — 숫자 / 데이터 / 전처리 / PCA
# ------------------------------------------------------------
import numpy as np                                # 누적합 계산
from sklearn.datasets import load_breast_cancer   # 유방암 데이터 (특성 30개)
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# ------------------------------------------------------------
# [목적] 스케일링 후 전체 주성분 구하고 -> 정보 누적합 -> 95% 넘는 축 수
# ------------------------------------------------------------
X = load_breast_cancer().data                     # 특성 30개 (=30차원)
X_s = StandardScaler().fit_transform(X)           # 분산 기준이라 스케일링 필수
pca = PCA().fit(X_s)                              # 30개 주성분 모두 구하기
cum = np.cumsum(pca.explained_variance_ratio_)    # 정보 누적 합 (1축→2축→…)
n95 = int(np.argmax(cum >= 0.95)) + 1             # 처음으로 95%를 넘는 지점
print('정보 95%를 지키는 데 필요한 축 수:', n95, '(전체 30개 중)')
```
