```python
# ############################################################
# 차원축소 예시 — PCA로 특성 줄이며 정보 지키기
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 + 데이터 준비 (와인 성분 13개 = 13차원)
# ------------------------------------------------------------
from sklearn.datasets import load_wine           # 와인 데이터 (특성 13개)
from sklearn.preprocessing import StandardScaler # 스케일링 (PCA는 분산 기준이라 필수)
from sklearn.decomposition import PCA            # 주성분분석 (특성 압축)
X = load_wine().data                             # 특성 13개
X_s = StandardScaler().fit_transform(X)          # 같은 잣대로 (퍼진 정도를 공정하게 재기)
```

```python
# ------------------------------------------------------------
# [목적] 13차원 -> 2차원으로 줄이고, 각 축이 담은 '정보량' 보기
# ------------------------------------------------------------
pca = PCA(n_components=2)                         # 2개(주성분)로 줄이기
X_2d = pca.fit_transform(X_s)                     # 13열 -> 2열로 압축
print('각 주성분이 담은 정보:', pca.explained_variance_ratio_.round(2))  # [1축, 2축]
print('합계:', round(pca.explained_variance_ratio_.sum(), 2))            # 2축이 지킨 정보 비율
```

```python
# ------------------------------------------------------------
# [목적] 개수 대신 '정보 95%를 남겨줘'라고 시키기 (축 수 자동 결정)
# ------------------------------------------------------------
pca = PCA(n_components=0.95)                      # 정보 95%를 지키는 선에서
X_r = pca.fit_transform(X_s)                      # 알아서 축 수를 정함
print('남은 컬럼 수:', X_r.shape[1], '(원래 13개)')  # 몇 축이면 95%가 되는지
```
