```python
# ############################################################
# Level 1 — 붓꽃 4차원 -> 2차원 (가장 단순)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기
# ------------------------------------------------------------
from sklearn.datasets import load_iris           # 붓꽃 데이터 (특성 4개)
from sklearn.decomposition import PCA            # 주성분분석
```

```python
# ------------------------------------------------------------
# [목적] 4차원 -> 2차원 압축 + 남긴 정보 확인
# ------------------------------------------------------------
X = load_iris().data                             # 특성 4개 (=4차원)
pca = PCA(n_components=2)                         # 2차원으로 줄이기
X_2d = pca.fit_transform(X)                       # 4열 -> 2열
print('줄인 뒤 모양:', X_2d.shape)                # (150, 2)
print('남긴 정보 비율:', pca.explained_variance_ratio_.round(2))  # [1축, 2축]
```

```python
# ============================================================
# [결과 해석]
#  · 남긴 정보 [0.92, 0.05] -> 두 축 합쳐 약 0.97 (원래 정보의 97%!)
#  · 4개 특성을 2개로 줄여도 정보 손실이 거의 없음 -> iris는 사실상 2차원으로 충분
#  · 이렇게 2차원이 되면 산점도로 '한눈에' 볼 수 있는 게 PCA의 큰 장점
# ============================================================
```
