```python
# ############################################################
# 군집화 예시 — 스케일링 & 무리 해석 (비지도)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기 — 데이터 / 군집 모델 / 전처리 준비
# ------------------------------------------------------------
import pandas as pd                               # 표 다루기 (엑셀 같은 표를 코드로 다루는 도구)
from sklearn.datasets import load_iris            # 붓꽃 데이터 (연습용으로 많이 쓰는 꽃 데이터)
from sklearn.cluster import KMeans                # K-means 군집화 (비슷한 것끼리 무리로 묶는 방법)
from sklearn.preprocessing import StandardScaler  # 스케일링 (숫자 크기를 공평하게 맞추기)
from sklearn.pipeline import make_pipeline        # 파이프라인 (여러 단계를 한 상자로 묶어 실행)
```

```python
# ------------------------------------------------------------
# [목적] 데이터 준비 — 붓꽃 특성 4개 (정답 품종은 사용 안 함 = 비지도)
# ------------------------------------------------------------
iris = load_iris(as_frame=True)                   # 표 형태로 로드 (행=꽃, 열=특성인 표로 받기)
X = iris.data                                     # 특성만 사용 (꽃받침·꽃잎의 길이/너비 4개)
X.head()                                           # 앞 5줄 미리보기 (데이터가 잘 들어왔는지 눈으로 확인)
```

```python
# ------------------------------------------------------------
# [목적] 스케일링 + K-means(3무리)를 파이프라인으로 한 번에
#        (스케일링을 빼면 숫자 큰 특성이 거리 계산을 지배함)
# ------------------------------------------------------------
model = make_pipeline(StandardScaler(),           # (1) 같은 잣대로 스케일링 (단위 차이를 없애 공평하게)
                      KMeans(n_clusters=3,         # (2) 3무리로 나누기 (몇 덩어리로 묶을지 정하기)
                             random_state=0,       #     결과 고정 (매번 같은 답이 나오게 씨앗값 고정)
                             n_init=10))           #     10번 시도 중 최선 (여러 번 해보고 제일 좋은 것 선택)
X['무리'] = model.fit_predict(iris.data)          # 무리 번호 부여 (각 꽃이 0/1/2 중 어디에 속하는지)
X['무리'].value_counts()                          # 무리별 인원 (각 덩어리에 꽃이 몇 송이인지)
```

```python
# ------------------------------------------------------------
# [목적] 무리별 평균으로 '성격' 파악 (의미·이름 붙이기는 사람 몫)
# ------------------------------------------------------------
X.groupby('무리').mean().round(2)                 # 무리별 특성 평균 (각 덩어리의 대표 크기를 평균으로 보기)
```
