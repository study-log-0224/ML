```python
# ############################################################
# Level 1 — 붓꽃 3무리로 나누기 (가장 단순)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기
# ------------------------------------------------------------
import numpy as np                                # 배열/개수 세기 (숫자 묶음을 빠르게 다루는 도구)
from sklearn.datasets import load_iris            # 붓꽃 데이터 (연습용 꽃 데이터)
from sklearn.cluster import KMeans                # K-means 군집화 (비슷한 것끼리 묶기)
```

```python
# ------------------------------------------------------------
# [목적] 데이터 준비 — 붓꽃 특성 4개 (정답 품종은 안 씀)
# ------------------------------------------------------------
data = load_iris()                                # 데이터 로드 (붓꽃 데이터 통째로 불러오기)
X = data.data                                     # 특성만 사용 (정답 품종은 빼고 크기 4개만)
print('데이터 크기:', X.shape)                    # 크기 확인 (꽃 150송이 × 특성 4개인지 보기)
```

```python
# ------------------------------------------------------------
# [목적] 학습 — K-means로 3무리 나누기
# ------------------------------------------------------------
km = KMeans(n_clusters=3,                         # 3무리로 (세 덩어리로 묶기)
            random_state=0,                       # 결과 고정 (매번 같은 답 나오게)
            n_init=10)                            # 10번 시도 중 최선 (여러 번 해보고 최고 선택)
labels = km.fit_predict(X)                        # 무리 번호(0/1/2) (각 꽃이 어느 덩어리인지 계산)

print('앞 10송이 무리:', labels[:10])             # 앞 10개 보기 (결과 맛보기)
print('무리별 개수:', np.bincount(labels))        # 무리별 인원 (덩어리마다 몇 송이인지)
```
