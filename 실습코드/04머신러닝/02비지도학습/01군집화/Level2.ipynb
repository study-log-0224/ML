```python
# ############################################################
# Level 2 — 와인 데이터 (특성 13개, 스케일링 + 실루엣)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기 — 데이터 / 전처리 / 모델 / 평가
# ------------------------------------------------------------
import numpy as np                                # 개수 세기 (숫자 묶음 다루기)
from sklearn.datasets import load_wine            # 와인 데이터 (성분이 담긴 연습 데이터)
from sklearn.preprocessing import StandardScaler  # 스케일링 (숫자 크기 공평하게 맞추기)
from sklearn.cluster import KMeans                # K-means (비슷한 것끼리 묶기)
from sklearn.metrics import silhouette_score      # 실루엣 점수 (무리가 잘 나뉘었는지 성적 매기기)
```

```python
# ------------------------------------------------------------
# [목적] 데이터 준비 — 와인 성분 13개 (단위 제각각: 알코올 12, 마그네슘 100대)
# ------------------------------------------------------------
X = load_wine().data                              # 특성 13개 (와인 성분 13가지)
print('데이터 크기:', X.shape)                    # 크기 확인 (와인 178병 × 성분 13개)
```

```python
# ------------------------------------------------------------
# [목적] 전처리 — 스케일링 (거리 기반이라 필수)
# ------------------------------------------------------------
X_s = StandardScaler().fit_transform(X)           # 스케일링 (성분마다 다른 단위를 같은 잣대로)
```

```python
# ------------------------------------------------------------
# [목적] 학습(3무리) + 평가(실루엣)
# ------------------------------------------------------------
km = KMeans(n_clusters=3, random_state=0, n_init=10)   # 3무리 모델 (세 덩어리로, 결과 고정)
labels = km.fit_predict(X_s)                      # 스케일한 값으로 학습 (원본 X가 아님에 주의!)

print('무리별 개수:', np.bincount(labels))        # 무리별 인원 (덩어리별 와인 수)
print('실루엣 점수:', round(silhouette_score(X_s, labels), 3))  # 품질 점수 (1에 가까울수록 깔끔하게 나뉨)
```
