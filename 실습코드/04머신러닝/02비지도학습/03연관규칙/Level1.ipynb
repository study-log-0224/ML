```python
# ############################################################
# Level 1 — 장바구니 규칙 (TransactionEncoder로 간단히)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기
# ------------------------------------------------------------
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder      # 거래목록 -> 0/1 표로 자동 변환
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# ------------------------------------------------------------
# [목적] 데이터 준비 + 0/1 표로 인코딩
# ------------------------------------------------------------
data = [['라면','계란'], ['라면','계란','콜라'], ['라면','콜라'], ['우유','빵'], ['라면','계란']]  # 거래 5건
te = TransactionEncoder()
df = pd.DataFrame(te.fit(data).transform(data), columns=te.columns_)   # 품목별 True/False 표
```

```python
# ------------------------------------------------------------
# [목적] 빈발 조합 -> 신뢰도 0.6 이상 규칙
# ------------------------------------------------------------
freq = apriori(df, min_support=0.4, use_colnames=True)             # 40% 이상 등장 조합
rules = association_rules(freq, metric='confidence', min_threshold=0.6)  # 신뢰도 60% 이상
print(rules[['antecedents','consequents','support','confidence','lift']].to_string())
```

```python
# ============================================================
# [결과 해석]
#  · TransactionEncoder가 거래목록을 자동으로 True/False 표로 바꿔줌 (수작업 불필요)
#  · 라면·계란·콜라가 자주 함께 등장 -> 관련 규칙이 뽑힘
#  · min_support/min_threshold를 낮추면 규칙이 많아지고, 높이면 확실한 것만 남음
# ============================================================
```
