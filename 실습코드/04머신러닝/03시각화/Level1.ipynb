```python
# ############################################################
# Level 1 — matplotlib 3종을 한 판에 (선·히스토그램·산점도)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 + 데이터 준비
# ------------------------------------------------------------
import numpy as np                                # 숫자·난수 만들기
import pandas as pd                               # 표(DataFrame)로 확인하기
import matplotlib.pyplot as plt                   # 그래프 그리기
%matplotlib inline
x = np.linspace(0, 10, 100)                       # 0~10을 100칸으로 (시간 흐름 흉내)
y = np.sin(x)                                     # 시간에 따른 변화(가정)
age = np.random.normal(35, 10, 500)               # 나이 분포(평균35, 퍼짐10)
ad  = np.random.uniform(100, 1000, 50)            # 광고비 50건 (100~1000 무작위)
sales = ad * 2.5 + np.random.normal(0, 500, 50)   # 광고비 <-> 매출 (비례 + 잡음)
```

```python
# ------------------------------------------------------------
# [데이터 살펴보기 · EDA] 그리기 전에 표로 확인 (광고비-매출)
# ------------------------------------------------------------
df = pd.DataFrame({'ad': ad, 'sales': sales})     # 광고비·매출을 표로 묶기
print('데이터 크기(행, 열):', df.shape)           # (50, 2) 확인
df.describe().round(1)                             # 요약 통계 (값 범위·평균 감 잡기)
```

```python
# ------------------------------------------------------------
# [목적] 그래프 3개를 나란히 그리기 (subplots)
# ------------------------------------------------------------
fig, ax = plt.subplots(1, 3, figsize=(13, 4))     # 1줄 3칸 도화지 (그래프 3개를 한 줄에)
ax[0].plot(x, y, color='orange');                 ax[0].set_title('Line')       # 선 (변화)
ax[1].hist(age, bins=20, color='skyblue', edgecolor='black'); ax[1].set_title('Histogram')  # 분포
ax[2].scatter(ad, sales, color='purple', alpha=0.6);          ax[2].set_title('Scatter')     # 관계
plt.tight_layout()                                # 그래프끼리 안 겹치게 간격 정리
plt.show()
```

```python
# ============================================================
# [결과 해석]
#  · 왼쪽(Line): 시간에 따라 오르내리는 흐름
#  · 가운데(Histogram): 나이가 35 근처에 몰린 종 모양 분포
#  · 오른쪽(Scatter): 광고비가 늘수록 매출도 느는 우상향(양의 관계)
#  · subplots로 여러 그래프를 한 화면에 나란히 비교할 수 있음
# ============================================================
```
