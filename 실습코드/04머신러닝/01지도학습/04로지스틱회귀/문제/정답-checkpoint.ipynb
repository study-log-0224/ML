```python
## 세포 측정치로 유방암 악성/양성 진단
## 타깃(라벨, y값) : target  ->  0(악성, malignant) / 1(양성, benign)
##
## 목적 : 세포 측정치로 종양이 악성인지 판정한다.
##
## 정할 것 4가지  (07 본편 health_survey와 비교)
##   1) 무엇을 맞히나  : target (악성/양성)
##   2) 답이 몇 종류인가 : 2종류 -> 이진분류   <- 본편과 같다
##   3) 양성(1)은?     : [주의] 여기서 사고가 난다. 아래 참고
##   4) 어떻게 채점하나 : 정확도 + ROC-AUC, 그런데 '악성을 놓친 수'를 반드시 볼 것
##
## 3)이 이 과제의 함정이다.
##
## 의학에서 '양성(benign)'은 '순한 혹, 괜찮다'는 뜻이다.
## 머신러닝에서 '양성(positive)'은 '우리가 찾고 싶은 쪽, 라벨 1'이라는 뜻이다.
## 같은 한국어 단어인데 뜻이 정반대다.
##
## 이 데이터에서 target=1 은 benign(순한 혹)이다.
## 즉 sklearn이 말하는 '양성 클래스(1)'는 '건강한 쪽'이다.
## 우리가 정작 찾고 싶은 건 악성(0)인데 말이다.
##
## 그래서 classification_report를 그냥 읽으면 뒤집혀 읽는다.
## recall이 높다고 좋아했는데 '건강한 사람을 잘 찾았다'는 뜻일 수 있다.
##
## '이 데이터에서 1이 뭐지?'를 시작할 때 확인하는 것 - 그게 01단계가 하는 일이다.
```

```python
# ----------------------------
# 02 데이터가져오기
# ----------------------------
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer(as_frame=True)
X = data.data          # 30개 세포 측정 특성
y = data.target        # 0=악성, 1=양성
print('X:', X.shape)   # (569, 30)
print('target_names:', list(data.target_names))   # ['malignant', 'benign']
#
# sklearn에 내장돼 있어 다운로드도 로그인도 필요 없다.
# target_names를 꼭 찍어 보자. 0번이 malignant(악성), 1번이 benign(양성)이다.
# 01단계에서 걱정한 그 함정이 여기서 확인된다.
```

```python
# ----------------------------
# 03 EDA
# ----------------------------
```

```python
print(y.value_counts())
#   1: 357  (양성 benign)
#   0: 212  (악성 malignant)
#
# 63:37. 본편(59:41)과 비슷하게 균형에 가깝다.
# 다수 클래스가 '건강한 쪽'이라는 게 상식과 맞는다.
```

```python
# 특성 범위를 보자 - 본편의 복선이 여기서 극단적으로 나타난다
X[['mean area', 'mean smoothness']].describe().loc[['min', 'max']].round(4)
#                mean area   mean smoothness
#   min             143.5            0.0526
#   max            2501.0            0.1634
#
# mean area는 143 ~ 2501, mean smoothness는 0.05 ~ 0.16.
# 범위가 약 15,000배 차이난다.
#
# 본편(health_survey)은 나이 18~79 vs 흡연 0~1로 약 40배 차이였다.
# 여기는 그것과 비교가 안 된다. 게다가 특성이 4개가 아니라 30개다.
# 이 차이가 08-2에서 본편과 다른 결과를 만든다.
```

```python
print('결측:', X.isnull().sum().sum())   # 0
#
# sklearn 내장 데이터는 이미 정제돼 있다. 04에서 할 일이 없다.
```

```python
# ----------------------------
# 04 데이터전처리
# ----------------------------
```

```python
# 할 게 없다 - 결측 0, 전부 숫자, 식별자 없음.
# 스케일링은 본편과 같은 이유로 05 분할 뒤에 한다(누수 방지).
print('X:', X.shape, '| y:', y.shape)   # (569, 30) (569,)
```

```python
# ----------------------------
# 05 검증데이터 분할 및 학습
# ----------------------------
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# 05-1) 분할
X_tr, X_val, y_tr, y_val = train_test_split(
    X, y, test_size=0.2, random_state=0, stratify=y)

# 05-2) 스케일링 - fit은 train에만
sc = StandardScaler()
X_tr_s = sc.fit_transform(X_tr)
X_val_s = sc.transform(X_val)

# 05-3) 학습
model = LogisticRegression(max_iter=5000).fit(X_tr_s, y_tr)
print(X_tr.shape, X_val.shape)   # (455, 30) (114, 30)
#
# max_iter=5000 - 기본값(100)으로는 수렴 경고가 뜰 수 있어 넉넉히 준다.
# 로지스틱은 정답을 한 번에 계산하는 게 아니라 조금씩 고쳐 가며 찾기 때문에,
# 특성이 많으면 반복 횟수가 더 필요하다.
```

```python
# ----------------------------
# 06 테스트 및 검증지표 확인
# ----------------------------
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, confusion_matrix

pred = model.predict(X_val_s)
proba = model.predict_proba(X_val_s)[:, 1]
print('정확도 :', round(accuracy_score(y_val, pred), 4))   # 0.9825
print('ROC-AUC:', round(roc_auc_score(y_val, proba), 4))   # 0.9957
print()
print(classification_report(y_val, pred, target_names=['악성(0)', '양성(1)']))
#
# 본편은 AUC 0.63이었는데 여기는 0.9957이다. 같은 로지스틱, 같은 코드인데.
# 모델이 똑똑해진 게 아니라 데이터가 다른 것이다.
# 세포 측정치는 악성/양성을 실제로 잘 가르는 정보를 담고 있다.
#
# target_names를 넘겨 주면 0/1 대신 이름이 찍혀 뒤집어 읽을 위험이 준다.
# 01단계에서 걱정한 그 함정에 대한 실전 대비책이다.
```

```python
# [중요] 정확도 0.98에 만족하고 넘어가면 안 된다. 혼동행렬을 봐야 한다.
cm = confusion_matrix(y_val, pred)
print(cm)
print()
print('악성인데 양성이라 놓친 수:', cm[0, 1])   # 2
print('양성인데 악성이라 오진한 수:', cm[1, 0])  # 0
#   [[40  2]
#    [ 0 72]]
#
# 하필 '악성을 놓친 쪽'으로만 2건 틀렸다. 반대 방향 실수는 0건이다.
#
# 이게 의료에서 가장 나쁜 방향의 실수다.
#   악성을 놓치면(FN) -> 환자가 치료를 못 받고 병이 진행된다
#   양성을 악성이라 하면(FP) -> 조직검사를 한 번 더 한다. 놀라지만 살 수 있다
#
# 정확도 98.25%는 훌륭해 보이지만, 그 1.75%가 전부 최악의 방향이다.
# 지표 하나만 보면 이걸 놓친다.
```

```python
# ----------------------------
# 07 Model 내보내기
# ----------------------------
```

```python
import joblib

joblib.dump(model, 'model_cancer.pkl')
joblib.dump(sc, 'scaler_cancer.pkl')   # 스케일러도 반드시 같이

lm = joblib.load('model_cancer.pkl')
ls = joblib.load('scaler_cancer.pkl')
print('재로드 예측 일치:', (lm.predict(ls.transform(X_val)) == pred).all())   # True
#
# 본편 07에서 본 그대로다. 모델만 저장하면 새 환자의 'mean area 1200'을
# 무엇으로 바꿔야 할지 알 수 없다.
# 여기는 범위 차이가 15,000배라 스케일러를 빠뜨리면 본편보다 훨씬 크게 틀린다.
```

```python
# ----------------------------
# 08 기타
# ----------------------------
```

```python
# 08-1) (문제 4번) 어느 실수가 더 위험한가 - 임계값으로 답한다
from sklearn.metrics import recall_score

for t in [0.5, 0.9]:
    pt = (proba >= t).astype(int)   # 양성(1)일 확률이 t 이상일 때만 '양성'
    c = confusion_matrix(y_val, pt)
    print('임계값 %.1f -> 놓친 악성 %d명 | 정확도 %.4f | 악성 recall %.4f'
          % (t, c[0, 1], accuracy_score(y_val, pt), recall_score(y_val, pt, pos_label=0)))
#   임계값 0.5 -> 놓친 악성 2명 | 정확도 0.9825 | 악성 recall 0.9524
#   임계값 0.9 -> 놓친 악성 1명 | 정확도 0.9386 | 악성 recall 0.9762
#
# 기준을 0.9로 올린다 = '양성(건강)이라고 말하려면 90% 이상 확신해야 한다'
# = 조금이라도 의심스러우면 악성으로 분류한다.
#
# 결과: 놓친 악성이 2명 -> 1명으로 줄었다. 대신 정확도는 0.98 -> 0.94로 떨어졌다.
#
# 정확도를 4%p 내주고 환자 한 명을 살렸다. 이게 남는 장사인가?
# 의료에서는 그렇다. 정확도는 숫자고 환자는 사람이다.
#
# 지표는 목적이 아니라 도구다. 01단계에서 '무엇이 중요한가'를 정하지 않으면
# 정확도가 높은 쪽을 고르게 되고, 그게 사람을 잡는다.
#
# pos_label=0 에 주의. 우리가 관심 있는 건 악성(0)의 recall이라
# sklearn의 기본값(1)을 뒤집어 줘야 한다. 01단계 3)의 함정이 코드로 나타난 것이다.
```

```python
# 08-2) 스케일링을 빼면? - 본편과 다른 결과가 나온다
import warnings
warnings.filterwarnings('ignore')

m_raw = LogisticRegression(max_iter=5000).fit(X_tr, y_tr)
print('[스케일링 O] 정확도 %.4f | AUC %.4f'
      % (accuracy_score(y_val, pred), roc_auc_score(y_val, proba)))
print('[스케일링 X] 정확도 %.4f | AUC %.4f'
      % (accuracy_score(y_val, m_raw.predict(X_val)), roc_auc_score(y_val, m_raw.predict_proba(X_val)[:, 1])))
#   [스케일링 O] 정확도 0.9825 | AUC 0.9957
#   [스케일링 X] 정확도 0.9474 | AUC 0.9924
#
# 본편(health_survey)에서는 스케일링을 빼도 정확도가 0.6150으로 똑같았다.
# 여기서는 0.9825 -> 0.9474로 확실히 떨어진다. 왜 다를까?
#
# 로지스틱에는 기본으로 규제(C=1.0)가 걸려 있다. 규제는 '계수를 너무 크게
# 만들지 마라'라고 누르는 장치다.
# 그런데 스케일을 안 맞추면 이 누르는 힘이 특성마다 불공평하게 걸린다.
#   mean area(143~2501)  : 계수가 작아도 되니 규제를 거의 안 받는다
#   mean smoothness(0.05~0.16) : 계수가 커야 하는데 규제에 눌려 버린다
# 즉 규제가 '범위가 작은 특성'만 골라서 죽인다.
#
# 본편은 특성 4개에 범위 차이 40배라 이 왜곡이 눈에 안 띄었다.
# 여기는 특성 30개에 범위 차이 15,000배라 성능으로 드러난 것이다.
#
# 교훈: '스케일링은 성능과 무관하다'는 절대 규칙이 아니다.
#       본편에서 그렇게 보였을 뿐이다. 범위 차이가 크고 특성이 많으면
#       성능에도 영향이 온다. 그러니 로지스틱에는 그냥 항상 하는 게 맞다.
```

```python
# 08-3) 계수로 해석 - 무엇이 악성을 가리키나
coef = pd.Series(model.coef_[0], index=X.columns)
print(coef.abs().sort_values(ascending=False).head(5).round(3))
#   radius error           1.325
#   worst radius           0.984
#   worst texture          0.973
#   mean concave points    0.963
#   worst area             0.867
#
# 크기·모양의 불규칙성이 상위에 온다. 상식(악성 종양은 크고 울퉁불퉁하다)과 맞는다.
#
# 절댓값으로 정렬한 이유는 부호가 '방향'일 뿐이기 때문이다.
# 여기서 부호를 읽을 땐 조심해야 한다 - 1이 '양성(benign, 건강)'이므로
# 계수가 음수여야 '악성 쪽으로 민다'는 뜻이 된다. 01단계 3)의 함정이 또 나온다.
#
# 이 비교가 가능한 것도 05-2에서 스케일링을 했기 때문이다.
# 안 했다면 mean area의 계수가 억울하게 작아져 순위가 엉망이 됐을 것이다.
```
