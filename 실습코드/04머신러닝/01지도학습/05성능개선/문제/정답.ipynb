> 목표: 교차검증으로 baseline을 확인하고, GridSearchCV로 랜덤포레스트를 튜닝해 성능을 끌어올린다.

## 1. 로드 & baseline (교차검증)

```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/winequality-red.csv')
y = df['quality']; X = df.drop(columns=['quality'])

base = cross_val_score(RandomForestRegressor(random_state=0), X, y, cv=5, scoring='r2').mean()
print('baseline 평균 R2:', round(base, 4))
```

## 2. GridSearchCV로 튜닝

```python
from sklearn.model_selection import GridSearchCV

params = {'n_estimators': [100, 300], 'max_depth': [None, 10, 20]}
grid = GridSearchCV(RandomForestRegressor(random_state=0), params, cv=5, scoring='r2')
grid.fit(X, y)
print('최적 파라미터:', grid.best_params_)
print('최적 CV R2   :', round(grid.best_score_, 4))
```

## 정리
- 교차검증으로 baseline을 잡고, GridSearchCV로 하이퍼파라미터를 튜닝해 R2를 개선.
- 08(당뇨)과 동일한 성능개선 흐름을 새 공공데이터에 적용.
- 튜닝은 시간이 걸린다(후보 x fold만큼 학습). 후보는 신중히 고른다.
