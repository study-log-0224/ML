## 풀이 6단계
1. 공공데이터 로드
```python
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
```
2. EDA - 타겟/결측
```python
df['Survived'].value_counts()
df[['Age', 'Embarked']].isnull().sum()
```
3. 특성 선택 & 전처리
```python
cols = ['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']
df = df[cols].copy()
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
X = pd.get_dummies(df.drop(columns=['Survived']))
```
4. 분할 & 학습
5. 평가 -> 정확도 0.7877, ROC-AUC 0.8195
6. 중요 특성 -> Sex, Fare, Age, Pclass 상위

## 해설 포인트
- 이름/티켓/객실번호(Cabin)는 고유값이 많거나 결측이 커서 그대로는 쓰기 어렵다 -> 제외(또는 가공).
- `Sex`가 가장 중요: 여성 생존율이 훨씬 높았다. `Pclass`/`Fare`: 상류층 생존율이 높았다.
- 04(인구조사)와 **동일한 워크플로우**가 새 데이터에 그대로 적용됨을 확인 -> 워크플로우의 재사용성.
